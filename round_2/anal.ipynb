```python
BaseMoney = 10000

# Extracted data from the image
data = [
    ("A", 10, 1),
    ("B", 80, 6),
    ("C", 37, 3),
    ("D", 17, 1),
    ("E", 31, 2),
    ("F", 90, 10),
    ("G", 50, 4),
    ("H", 20, 2),
    ("I", 73, 4),
    ("J", 89, 8)
]

print("Sensitivity (10 - Inhabitants):")
for name, multiplier, inhabitants in data:
    print(f"{name}: {10 - inhabitants}")

def printPayouts(percentChoose):
    print(f"\nPayouts with {percentChoose}% extra load:")
    for name, multiplier, inhabitants in data:
        payout = BaseMoney * multiplier / (inhabitants + percentChoose)
        print(f"{name}: {payout:.2f}")

print("\nMoney with no percent added:")
printPayouts(0)

print("\nMoney with average percent, 1% each:")
printPayouts(1)

print("\nMoney with average percent, 2.5% each:")
printPayouts(2.5)

print("\nMoney with average percent, 4% each:")
printPayouts(4)

print("\nMoney with average percent, 8% each:")
printPayouts(8)

print("\nMoney with average percent, 10% each:")
printPayouts(10)

print("\nMoney with max percent (everyone uses all expeditions):")
printPayouts(33)

print("\nMoney with max percent (everyone uses one expedition):")
printPayouts(100)

# Sort and rank
ranked = sorted(
    [(name, multiplier, BaseMoney * multiplier / inhabitants) for name, multiplier, inhabitants in data],
    key=lambda x: x[2],
    reverse=True
)

print("\n--- Ranked by Base Payout (no percent) ---")
for item in ranked:
    print(f"{item[0]}: Multiplier={item[1]}, Payout={item[2]:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [
    ("A", 10, 1),
    ("B", 80, 6),
    ("C", 37, 3),
    ("D", 17, 1),
    ("E", 31, 2),
    ("F", 90, 10),
    ("G", 50, 4),
    ("H", 20, 2),
    ("I", 73, 4),
    ("J", 89, 8)
]

for item in data : 
    d = np.array([item[1]/(item[2]+percent) for percent in range (0,101)])

    plt.plot(d,label=f"{item[1]} -> {item[2]}")
plt.xlabel("percent")
plt.ylabel("multiplier")
plt.legend()
plt.show()
```
