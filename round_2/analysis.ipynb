```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np  
```

```python
day1 = "/Users/sarthak/Desktop/imc/round_2/round-2-island-data-bottle/prices_round_2_day_-1.csv"
day2 = "/Users/sarthak/Desktop/imc/round_2/round-2-island-data-bottle/prices_round_2_day_0.csv"
day3 = "/Users/sarthak/Desktop/imc/round_2/round-2-island-data-bottle/prices_round_2_day_1.csv"
```

```python
day_1 = pd.read_csv(day1,delimiter=';')
day_2 = pd.read_csv(day2,delimiter=';')
day_3 = pd.read_csv(day3,delimiter=';')
```

```python
final = pd.concat([day_1,day_2,day_3],axis=0)
```

```python
final
```

```python
croissant = final.loc[final['product'] == 'CROISSANTS']
djembes = final.loc[final['product'] == 'DJEMBES']
jams = final.loc[final['product'] == 'JAMS']
b1 = final.loc[final['product'] == 'PICNIC_BASKET1']
b2 = final.loc[final['product'] == 'PICNIC_BASKET2']
```

```python
pb_1 = 6*croissant.mid_price.values + 1*djembes.mid_price.values + 3*jams.mid_price.values 
pb_2 = 4*croissant.mid_price.values + 2*jams.mid_price.values
```

```python
spread_1 = b1.mid_price.values - pb_1
plt.plot(spread_1)
print(spread_1.mean())
```

```python
spread_2 = b2.mid_price.values - pb_2
plt.plot(spread_2)
print(spread_2.mean())
```
